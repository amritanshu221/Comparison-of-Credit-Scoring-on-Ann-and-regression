# One hot encoding for categorical columns

```python
# to seperate "Geography" column into three columns based on different country
df1 = pd.get_dummies(data = df, columns = ['Geography'])
df1.head()
```

```python
# to normalize the range of some columns for easy prediction 
scaled_col = ['CreditScore','Age','Balance','EstimatedSalary']
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
df1[scaled_col] = scalar.fit_transform(df1[scaled_col])
```

```python
df1.head()
```

```python
X = df1.drop('Exited',axis = 'columns')
y = df1['Exited']
```

```python
# to randomly split train and test data into two parts
from sklearn.model_selection  import train_test_split
X_train,X_test , y_train ,y_test = train_test_split(X , y , test_size= 0.2, random_state = 5)
```

# Build ANN Model

```python

def model(X_train, X_test, y_train, y_test):
  model=  keras.Sequential([
      keras.layers.Dense(600, input_shape=(len(X_test.columns),), activation = 'relu'),
      keras.layers.Dense(300, activation = 'relu'),
      keras.layers.Dense(1, activation = 'sigmoid')

  ])
  # to compile the model
  model.compile(optimizer = 'adam',loss = 'binary_crossentropy',metrics=['accuracy'])

  # Model fitting is the measure of how well a machine learning model generalizes data similar to that with which it was trained.
  model.fit(X_train,y_train, epochs = 100)
  print(f'\n test accuracy -> {model.evaluate(X_test, y_test)}')

  y_pred = model.predict(X_test)

    
  return y_pred
```

```python
# predicting the credit score 
y_pred = model(X_train, X_test, y_train,y_test)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
 # converting predicted value into binary  
yp = []
for element in y_pred:
    if element > 0.5:
        yp.append(1)
    else:
        yp.append(0)
```

# confusion matrix

```python
# A confusion matrix is a table that is used to define the performance of a classification algorithm
# to compare actual vs predicted credit score
import seaborn as sns 
cm = tf.math.confusion_matrix(labels=y_test, predictions=yp)
plt.figure(figsize=(8,5))
sns.heatmap(cm , annot=True,fmt = 'd')
plt.xlabel("PREDICTION")
plt.ylabel("ACTUAL")
plt.show()
```
